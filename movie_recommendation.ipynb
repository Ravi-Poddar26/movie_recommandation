```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
['loved','loving','love']
['love', 'love', 'love']
```

```python
ps.stem('danced')
```

```python
# This cell caused a SyntaxError due to an unescaped apostrophe in the string literal.
# It's generally better to apply stemming to each word after splitting a sentence, which the 'stem' function already does.
# Therefore, we will apply the defined 'stem' function to the 'tags' column directly.
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
#vectorization by using bags of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
print(list(cv.get_feature_names_out()))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(movie):
  # Convert the input movie title and the DataFrame's title column to lowercase for case-insensitive matching
  movie_index = new_movies[new_movies['title'].str.lower() == movie.lower()].index[0]
  # Correctly call the cosine_similarity function
  similarities_for_movie = cosine_similarity(vectors[movie_index].reshape(1, -1), vectors)
  # The result is a 1xN array, so we need to access its first (and only) row
  distances = similarities_for_movie[0]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_movies.iloc[i[0]].title)
```

```python
recommend('jumanji')
```
